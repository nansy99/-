#### 7. Напишите функцию sorted_degree(). 
Функция сортирует пары (index_node, degree) по значению степени в убывающем порядке. 
Выведите топ-10 вершин с наибольшей степенью. 

```python
def sorted_degree(G):
    return sorted(G.degree, key=lambda x: x[1], reverse=True)

for i in range(0,10):
    print(sorted_degree(G)[i])
```

#### 11. Сколько процентов узлов из имеющих максимальную кластеризацию имеют степень два (т.е. являются вершинами треугольников)? (ответ округлите до целого)

```python
count_max_cluster=0
count_2= 0 
max_clustering = max(nx.clustering(G).values())
for node,d in G.degree():
    if (nx.clustering(G,node) == max_clustering):
        count_max_cluster+=1
for node,d in G.degree():
    if (d==2) and (nx.clustering(G,node) == max_clustering):
        count_2+=1
print(int((count_2/count_max_cluster)*100), "%")
```

#### 12. Какова максимальная степень вершины с  минимальной кластеризацией ($C_i=0$)?

```python
d_max= 0
for node,d in G.degree():
    if (nx.clustering(G,node) == 0):
        if (d > d_max):
            d_max = d
print(d_max)
```

#### 13. Какова кластеризация хаба (вершины с максимальной степенью)? (ответ округлите до сотых)

```python
degrees=[d for node, d in G.degree()]
d_max = max(degrees)
for node,d in G.degree():
    if (d == d_max):
        print(round(nx.clustering(G,node),2))
```

#### 19. Между какой парой вершин (i,j) ($i\ne j$) число путей длины 4 наибольшее? 

```python
A = nx.adjacency_matrix(G)
path_4 = A.dot(A.dot(A.dot(A))).todense()
```

```python
path_4.shape[0]
```

```python
for i in range(path_4.shape[0]):
    path_4[i,i] = 0
path_4
```

```python
result = np.where(path_4 == np.amax(path_4))
listOfCordinates = list(zip(result[0], result[1]))
for cord in listOfCordinates:
    print(cord)
```
