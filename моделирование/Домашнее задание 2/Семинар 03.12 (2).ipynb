# Случайные графы Эрдеша-Реньи

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx
```

### Генератор G(N,p). Распределение по степеням связности
Напишем свой генератор случайного графа Эрдеша-Реньи: 
каждую пару вершин (P,Q) с вероятностью $ \bf p$ соединяем ребром. 
Конечно, можно воспользоваться и готовым [erdos_renyi_graph.](https://networkx.github.io/documentation/networkx-1.10/reference/generated/networkx.generators.random_graphs.erdos_renyi_graph.html#networkx.generators.random_graphs.erdos_renyi_graph)

```python
def ER_graph(N,p):
    G=nx.Graph()
    G.add_nodes_from(np.arange(N))
    for P in range(N):
        for Q in range (P+1,N):
            if (np.random.random()<p):
                G.add_edge(P,Q)
    return G
```

```python
N,p=1000,0.005
G=ER_graph(N,p)
print ('Number of edges =',G.number_of_edges())
degrees=[d for node, d in G.degree()]
print ('Average degree =',np.mean(degrees))
```

```python
def plot_degree_distribution(degrees):   
    rho=np.bincount(degrees)/len(degrees)
    k=np.arange(len(degrees))[np.nonzero(rho)]
    rho_k=rho[np.nonzero(rho)]

    plt.plot(k,rho_k,'--o')
    plt.xlabel(r'$ k$', fontsize=14)
    plt.ylabel(r'$\rho(k)$', fontsize=14)
```

```python
plot_degree_distribution(degrees)
```

```python
degrees_rand=np.random.poisson(N*p,1000)
plot_degree_distribution(degrees)
plot_degree_distribution(degrees_rand)
```

## Эволюция сети Эрдеша-Реньи 

```python
N=100
prob=[0.005,0.01,0.025,0.05]
fig, axes = plt.subplots(nrows=2, ncols=2,figsize=(12, 8))
ax=plt.subplot(221)
G=ER_graph(N,prob[0])
nx.draw(G,nx.spring_layout(G,iterations=10),node_size=25,with_labels=False)
ax.set_title("p="+str(prob[0]))

ax=plt.subplot(222)
G=ER_graph(N,prob[1])
nx.draw(G,nx.spring_layout(G,iterations=10),node_size=25,with_labels=False)
ax.set_title("p="+str(prob[1]))

ax=plt.subplot(223)
G=ER_graph(N,prob[2])
nx.draw(G,nx.spring_layout(G,iterations=10),node_size=25,with_labels=False)
ax.set_title("p="+str(prob[2]))

ax=plt.subplot(224)
G=ER_graph(N,prob[3])
nx.draw(G,nx.spring_layout(G,iterations=10),node_size=25,with_labels=False)
ax.set_title("p="+str(prob[3]))
```

Построим зависимость размера гигантской компоненты от вероятности $\bf p$.  Для этого построим ансамбли случайных графов: 
для каждого значения $\bf p$
* сгенерируем $N_{ens}$ реализаций случайного графа из заданного распределения;
* вычислим размер гигантской компоненты для каждого графа из ансамбля;
* определим среднее значение и среднеквадратичное отклонение для размера гигантской компоненты.

Для построения зависимости $С(\bf p)$ воспользуемся функцией [errorbar.](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.errorbar.html)

```python
N,N_Ens=100,100
prob=np.linspace(0.0,0.05,101)
ensemble=np.zeros([101,N_Ens])
for i in range(len(prob)):
    for j in range(N_Ens):
        G=ER_graph(N,prob[i])
        Gc = max(nx.connected_components(G), key = len)
        ensemble[i,j]=len(Gc)/N
```

```python
plt.errorbar(prob,np.mean(ensemble,axis=1),np.std(ensemble,axis=1))
plt.xlabel(r'$ p$', fontsize=14)
plt.ylabel(r'$ GC$', fontsize=14)
```
